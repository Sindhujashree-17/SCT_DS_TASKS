```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/world_population_data.csv")
```

```python
print("First 5 rows of the dataset:")
print(df.head())
```

```python
print("\nDataset Information:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
missing_values = df.isnull().sum()
print("\nMissing Values:\n", missing_values)
```

```python
print("\nMissing Values After Cleaning:\n", df.isnull().sum())
```

```python
continent_population = df.groupby('continent')['2023 population'].sum()
plt.figure(figsize=(10, 6))
plt.bar(continent_population.index, continent_population.values, color='skyblue')
plt.title('Population Distribution by Continent (2023)', fontsize=14)
plt.xlabel('Continent', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['2023 population'], bins=10, color='orange', edgecolor='black')
plt.title('Population Size Distribution (2023)', fontsize=14)
plt.xlabel('Population (2023)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```
